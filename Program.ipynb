```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[19,49,67,45,23,32]
```

```python
plt.plot(x,y,color='green',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=12)
plt.title('Line graph')
plt.xlabel("Reg no")
plt.ylabel('Marks')
plt.show()
```

```python
z=[1,3,4,5,2,6]
```

```python
plt.plot(x,y,label='Marks')
plt.plot(x,z,label='Rank')
plt.ylabel("Rank and Marks")
plt.xlabel("Reg no")
plt.title("Line Graph")
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'g*',linestyle='dotted')
plt.xlabel("X-AXIS")
plt.ylabel("Y-LABEL")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y)
```
